```python
import random, math

num_of_cities = 52
num_of_individuals = 100

def save_result(e):
    f = open('result.txt', 'w')
    list_of_cities = list()
    for i in range(1, num_of_cities + 1):
        list_of_cities.append(i)
    for i in e:
        s = list_of_cities.pop(i - 1)
        f.write(str(s) + '\n')

cities_coord = dict()

def read_cities():
    f = open('berlin.txt', 'r')
    lines = f.readlines()
    for line in lines:
        line_list = [float(i) for i in line.split(" ")]
        cities_coord[line_list[0]] = line_list[1], line_list[2]
        
def f(e):
    list_of_cities = list()
    for i in range(1, num_of_cities + 1):
        list_of_cities.append(i)
    value_sum = 0
    for i in range(int(len(e) - 1)):
        start_point = cities_coord[list_of_cities.pop(e[i] - 1)]
        end_point = cities_coord[list_of_cities[e[i + 1] - 1]]
        dx = end_point[0] - start_point[0]
        dy = end_point[1] - start_point[1]
        value = math.sqrt(dx * dx + dy * dy)
        value_sum += value
    return value_sum

def selection(population):
    parents = list()
    for i in range(num_of_individuals):
        index1 = random.randint(0, num_of_individuals - 1)
        index2 = random.randint(0, num_of_individuals - 1)
        if(f(population[index1]) < f(population[index2])):
            parents.append(population[index1])
        else:
            parents.append(population[index2])
    return parents
    
def crossingover(parents):
    new_population = list()
    for i in range(int(len(parents)/2)):
        e_first = parents[i * 2]
        e_second = parents[i * 2 + 1]
        e_new_1 = []
        e_new_2 = []
        for t in range(len(e_first)):
            a = e_first[t]
            b = e_second[t]
            swapMask = a ^ b
            a = a ^ swapMask
            b = b ^ swapMask
            e_new_1.append(a)
            e_new_2.append(b)
            
        new_population.append(e_new_1)
        new_population.append(e_new_2)
    return new_population
    
def mutation(population):
    for i in range(len(population)):
        k = random.randint(0, 51)
        j = random.randint(0, 52 - k)
        new_ind = population[i]
        new_ind[k] = j
        if(f(new_ind) < f(population[i])):
            population[i][k] = j
    return population
    
def get_best(population):
    min = f(population[0])
    best_e = population[0]
    for e in population:
        if(f(e) < min):
            min = f(e)
            best_e = e
    return best_e

def main():
    first_population = list()
    read_cities()
    for j in range(num_of_individuals):
        e = []
        for i in range(num_of_cities):
            e.append(random.randint(1, num_of_cities - i))
        first_population.append(e)
        
    population = first_population
    best_result = f(get_best(first_population))
    best_population = first_population
    for i in range(1000):
        population = selection(population)
        population = crossingover(population)
        population = mutation(population)
        
        if(f(get_best(population)) < best_result):
            best_result = f(get_best(population))
            best_population = population
            
    save_result(get_best(best_population))
    print(best_result)
    
main()
```
