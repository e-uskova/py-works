```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def A(x):
#     return 2 * x - (1 / 6) * x ** 2
    return (6 - x) / 3
```

```python
def B1(x):
#     return (1 / 4) * x ** 2 - (3 / 2) * x
    return (x - 3) / 2 
```

```python
def B2(x):
#     return (7 / 2) * x - (1 / 4) * x ** 2
    return (7 - x) / 2
```

```python
xa = np.linspace(3, 6, 30)
xb1 = np.linspace(3, 5, 20)
xb2 = np.linspace(5, 7, 20)
plt.plot(xa, A(xa), label='A')
plt.plot(xb1, B1(xb1), label='B')
plt.plot(xb2, B2(xb2), label='B')
plt.grid()
plt.legend()
plt.title("Alpha-levels")
plt.savefig("alpha-levels.png")
```

```python
def suma(x):
    return 9 - x, 13 - 5 * x
```

```python
def sub(x):
    return - 5 * x + 3, x - 1
```

```python
def mul(x):
    return - 6 * x ** 2 + 3 * x + 18, 6 * x ** 2 - 33 * x + 42
```

```python
def div(x):
    return (6 - 3 * x) / (2 * x + 3), (6 - 3 * x) / (7 - 2 * x)
```

```python
alpha = np.array([0, 0.2, 0.4, 0.6, 0.8, 1])
s = suma(alpha)
su = sub(alpha)
m = mul(alpha)
d = div(alpha)
```
