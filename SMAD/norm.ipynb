```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from math import pi
```

```python
def func(x, a, sigma):
    return 1 / (sigma * np.sqrt(2 * pi)) * np.exp(-((x - a) ** 2) / (2 * sigma ** 2))
```

```python
X = np.array([2, 5, 7, 9, 12, 14, 17])
N = np.array([4, 7, 17, 31, 21, 14, 6])
p = N / np.sum(N)
M = np.sum(X * p)
M2 = np.sum(X ** 2 * p)
D = M2 - M ** 2
sigma = np.sqrt(D)
```

```python
plt.scatter(X, p, marker='o', label='Данные')
x = np.linspace(np.min(X), np.max(X), 30)
plt.plot(x, func(x, M, sigma), label='Норм.распр.')
plt.vlines(M, 0, 0.35, color='green', label='Мат.ожидание')
plt.grid()
plt.legend()
text = ('Дано:\nX = ' + str(list(X)) + '\nN = ' + str(list(N)) + '\nПолучено:\n' + 
       'a = ' + str(round(M, 3)) + '\nsigma = ' + str(round(sigma, 3)))
plt.text(1.8, 0.23, s=text, backgroundcolor='white')
# plt.savefig("norm.png")
```
