```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageColor
```

```python
graph = {1: {4, 8}, 2: {1, 6}, 3: {5}, 4: {8}, 5: {2}, 6: {3, 5, 8}, 7: {4}, 8: {9}, 9: {7}}
```

```python
graph = {}
f = open('graph640k.txt', 'r')
for line in f:
    line = line.strip().split(': ')
    graph[int(line[0])] = eval(line[1])
# graph
```

```python
i_graph = dict.fromkeys(graph.keys(), 0) 
for k, v in graph.items():
    for vertice in v:
        if i_graph[vertice]:
            i_graph[vertice].add(int(k))
        else:
            i_graph[vertice] = {int(k)}
for k, v in i_graph.items():
    if v == 0:
        i_graph[k] = set()
```

```python
n = len(graph)
in_time = [0] * (n + 1)
out_time = [0] * (n + 1)
cur_time = 0
```

```python
n
```

```python
def dfs(gr, v, comp, visited):
    comp.append(v)
#     global cur_time
#     in_time[v] = cur_time
    visited[v] = True
    for e in gr[v]:
        if not visited[e]:
#             cur_time += 1
            dfs(gr, e, comp, visited)
#     cur_time += 1
#     out_time[v] = cur_time
```

```python
visited = [False] * (n + 1)
cur_time = 0

coms = []

for i in range(1, n + 1):
    if not visited[i]:
        comp = []
#         cur_time += 1
        dfs(i_graph, i, comp, visited)
        coms.append(comp)
```

```python
print(coms)
```

```python
visited = [False] * (n + 1)
c = []

for comp in coms[::-1]:
    for v in comp[::-1]:
        if not visited[v]:
            co = []
            dfs(graph, v, co, visited)
            if len(co) > 1:
                c.append(co)
```

```python
width, height = 800, 800
```

```python
image = Image.new("RGB", (width, height))
```

```python
draw = ImageDraw.Draw(image)
```

```python
def scale(x, y):
    return (x + 2) * 200 , height - (y + 2) * 200
```

```python
x_0, y_0 = -2, 2
d = 0.005
N = int(4 / d)
N
```

```python
def fill_square(number, d, color):
#     print(color)
    y, x = divmod(number - 1, N)
#     x = x - 1
    x, y = x_0 + x * d, y_0 - y * d
    draw.rectangle((*scale(x, y), *scale(x + d, y - d)), fill=color)
```

```python
%%time
k = 0
r, g, b = 0, 0, 0
colors = [(255, 0, 0), (255, 120, 0), (255, 255, 0), (0, 255, 0), (0, 255, 255), (0, 0, 255), (120, 0, 255), (255, 0, 255)]
colors = colors * divmod(len(c), len(colors))[0] + colors[:divmod(len(c), len(colors))[1]]
for color, comp in zip(colors, c):
    r, g, b = 0, 255, 255
    if len(comp) > 20:
        k += 1 
        for v in comp:
            fill_square(v, d, (r, g, b))
```
