```python
num_training = 5000
indx = range(num_training)
X_train = X_train.iloc[indx].astype('float')
y_train = y_train.iloc[indx].astype('int')

num_test = 500
indx = range(num_test)
X_test = X_test.iloc[indx].astype('float')
y_test = y_test.iloc[indx].astype('int')
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], -1))
X_test = np.reshape(X_test, (X_test.shape[0], -1))
print(X_train.shape, X_test.shape)
```

```python
class KNeNe:
    
    def __init__(self):
        pass
    
    def train(self, X, y):
        self.X_train = X
        self.y_train = y
        
    def predict(self, X, k=1):
        dist = self.compute_distances(X)
        return self.predict_labels(dist, k=k)
    
    def compute_distances(self, X):
        num_test = X.shape[0]
        num_train = self.X_train.shape[0]
        dists = np.zeros((num_test, num_train))
        for i in range(num_test):
            dist[i,:] = np.sqrt(np.sum(np.square(self.X_train - X[i,:]), axis = 1))
        return dists
        
    def predict_labels(self, distd, k=1):
        num_test = dists.shape[0]
        y_pred = np.zeros(num_test)
        for i in range(num_test):
            idx = np.argsort(dists[i,:])[:k]
            neighbors = self.y_train[idx].ravel()
            cntr = Counter(neighbors)
            y_pred[i] = cntr.most_common(1)[0][0]
            
        return y_pred
```

```python
clf = KNeNe()
clf.train(X_train, y_train)
```

```python
dists = clf.compute_distances(X_test)
```

```python
y_test_pred = clf.predict_labels(dists, k=1)
accuracy = np.mean(y_test.ravel() == y_test_pred)
print('accuracy: %f'%accuracy)
```
