```python
import numpy as np
from scipy.stats import norm, chi2, t
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('whitegrid')

def pdf_plot(data, pdf, title):
    sns.lineplot(x = data, y = pdf, color = 'black')
    plt.xlabel('Data points')
    plt.ylabel('Probability Density')
    plt.title(title)
    plt.show()
```

```python
mu = 12
sigma = 2.5

data = np.arange(0, 24, 0.01)
n = norm(loc = mu, scale = sigma)
pdf = n.pdf(data)

pdf_plot(data, pdf, 'Нормальное распределение с μ={} и σ={}'.format(mu, sigma))

print('p(ξ < 0) = ' + 
      str(n.cdf(0)))
print('p(ξ > 0) = ' + 
      str(1 - n.cdf(0)))
print('p(ξ < μ) = ' + 
      str(n.cdf(mu)))
print('p(ξ > μ) = ' + 
      str(1 - n.cdf(mu)))
print('p(μ - σ < ξ < μ + σ) = ' + 
      str(n.cdf(mu + sigma) - n.cdf(mu - sigma)))
print('p(μ - 2σ < ξ < μ + 2σ) = ' + 
      str(n.cdf(mu + 2 * sigma) - n.cdf(mu - 2 * sigma)))
print('p(μ - 3σ < ξ < μ + 3σ) = ' + 
      str(n.cdf(mu + 3 * sigma) - n.cdf(mu - 3 * sigma)))
```

```python
data = np.arange(0, 24, 0.01)

df = 1

c = chi2(df = df)
pdf = c.pdf(data)

pdf_plot(data, pdf, 'Распределение χ2 с {} степенями свободы'.format(df))

print('p(χ2 < 0) = ' + 
      str(c.cdf(0)))
print('p(χ2 > 0) = ' + 
      str(1 - c.cdf(0)))
print('p(χ2 < n) = ' + 
      str(c.cdf(df)))
print('p(χ2 > n) = ' + 
      str(1 - c.cdf(df)))
```
