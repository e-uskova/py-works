```python
import numpy as np
from scipy.stats import norm, expon, chi2, t, uniform
import matplotlib.pyplot as plt
%matplotlib inline
```

Количество шагов за день средним значением 5000 и стандартным отклонением 2000 в течение года

```python
mu = 5000
sigma = 2000
n = 365
data = np.random.normal(mu, sigma, n)
x = np.arange(min(data), max(data), 2)
```

```python
_, bins, _ = plt.hist(data, 20, color = 'b', alpha=0.7)
pdf = norm.pdf(x, mu, sigma) * n * np.diff(bins)[0]
plt.plot(x, pdf, lw=2, color = 'r')
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
data_mu = np.mean(data)
print('μ = {}'.format(data_mu))
data_d = np.var(data)
print('D = {}'.format(data_d))
data_sigma = np.sqrt(data_d)
print('σ = {}'.format(data_sigma))
```

```python
beta = 1 / 3    # 1 / lambda
n = 300
data = np.random.exponential(beta, n)
x = np.arange(min(data), max(data) + 1, 0.1)
```

```python
_, bins, _ = plt.hist(data, 20, color = 'b', alpha=0.7)
pdf = expon.pdf(x, scale = beta) * n * np.diff(bins)[0]
plt.plot(x, pdf, lw=2, color = 'r')
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
data_mu = np.mean(data)
print('μ = {}'.format(data_mu))
data_d = np.var(data)
print('D = {}'.format(data_d))
data_sigma = np.sqrt(data_d)
print('σ = {}'.format(data_sigma))
```

```python
a = -5
b = 10
n = 300
data = np.random.uniform(a, b, n)
x = np.arange(min(data), max(data) + 1)
```

```python
_, bins, _ = plt.hist(data, 20, color = 'b', alpha=0.7)
pdf = uniform.pdf(x, a, b - a + 1) * n * np.diff(bins)[0]
plt.plot(x, pdf, lw=2, color = 'r')
plt.show()
```

```python
plt.boxplot(data)
plt.show()
```

```python
data_mu = np.mean(data)
print('μ = {}'.format(data_mu))
data_d = np.var(data)
print('D = {}'.format(data_d))
data_sigma = np.sqrt(data_d)
print('σ = {}'.format(data_sigma))
```
