```python
import numpy as np
from math import pi, sin
import matplotlib.pyplot as plt
import scipy.signal
```

```python
f1 = 8
f2 = 12
f3 = 24
# fsample1 = 96
fsample1 = 40
```

```python
secs = 5
t = np.arange(0, secs, 1/fsample1)
```

```python
a = np.array([sin(2 * pi * f1 * ti) for ti in t])
b = np.array([sin(2 * pi * f2 * ti) for ti in t])
c = np.array([sin(2 * pi * f3 * ti) for ti in t])
```

```python
f = np.arange( 0, fsample1, fsample1/len(t))
len(f)
```

```python
xa = abs(np.fft.fft(a))
xb = abs(np.fft.fft(b))
xc = abs(np.fft.fft(c))
```

```python
N = len(t)
N
```

```python
plt.plot(f, xa, 'red', f, xb, 'blue', f, xc, 'green')
plt.title('')
plt.show()
```

```python
d = a + 0.75 * b + 0.5 * c
AmpSpec = (1 / (N / 2)) * abs(np.fft.fft(d))
```

```python
plt.plot(d, linewidth=1)
plt.show()
```

```python
plt.plot(AmpSpec)
plt.show()
```

```python
w = scipy.signal.hamming(N)
dw = d * w
```

```python
plt.plot(dw, 'orange')
plt.show()
```

```python
AmplSpec_dw = abs(np.fft.fft(dw))
plt.plot(AmplSpec_dw, 'red', linewidth=1)
plt.show()
```
